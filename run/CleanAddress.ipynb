```python
import pandas as pd
import json

# Load the JSON data (assuming you have a JSON file)
with open('../assets/arrVietnam.json', 'r', encoding='utf-8') as f:
    arr_vietnam = json.load(f)

# Extract the 'value' field from each object and push to a list
value_list = [item['value'] for item in arr_vietnam]

# Print the result
print(value_list)
```

```python
# Convert value_list to a DataFrame
df_values = pd.DataFrame(value_list, columns=['City Values'])

# Define the output file path
output_file_path = '../assets/Value_List.xlsx'

# Export the DataFrame to Excel
df_values.to_excel(output_file_path, index=False, sheet_name='Values', engine='openpyxl')

# Print a message indicating the file has been saved
print(f"Value list exported to {output_file_path}")
```

```python
# Iterate over each value in value_list
for value in value_list:
    # Check whether the current value from value_list is present as a substring in each cell of the "city" column, ignoring case
    mask = df['city'].str.contains(value, na=False, case=False)

    # Replace the matching cells with the value (case-insensitive)
    df.loc[mask, 'city'] = value

# Display the updated DataFrame
print(df)
```

```python
# Define the output file path
output_file_path = '../assets/Updated_Customer.xlsx'

# Export the DataFrame to Excel
df.to_excel(output_file_path, index=False, sheet_name='Customers', engine='openpyxl')

# Print a message indicating the file has been saved
print(f"DataFrame exported to {output_file_path}")
```
